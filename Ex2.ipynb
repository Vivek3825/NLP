```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
# Download resources
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Sample text
text = "Natural Language Processing allows machines to understand human language."
```

```python
# Tokenize words
words = word_tokenize(text)
```

```python
# Apply POS tagging
pos_tags = nltk.pos_tag(words)
```

```python
# Display the result
print("POS Tags:", pos_tags)
```

## Student Activity:
1. Implement POS tagging on a paragraph related to any subject in your curriculum.

```python
newText = ("Part-of-Speech tagging (POS tagging) is the process of labeling each word "
            "in a sentence with its grammatical category (like noun, verb, adjective, etc.)."
            "It helps machines understand the role of each word in a sentence.")
```

```python
words = word_tokenize(newText)
pos_tags = nltk.pos_tag(words)
print("POS tags: ", pos_tags)
```

2. Analyze and count the frequency of different parts of speech (e.g., nouns, verbs).

```python
pos_tags = [tag for (word, tag) in pos_tags]
pos_counts = Counter(pos_tags)
```

```python
print("Part of Speech Frequency:")
for pos, count in pos_counts.items():
    print(f"{pos}: {count}")
```
