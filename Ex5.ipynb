```python
# Import required libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
```

```python
# Sample data (replace with your dataset)
corpus = [
    "Natural language processing enables computers to understand human language",
    "Word embeddings like Word2Vec represent words in vector space",
    "Bag of words and TF-IDF are traditional NLP approaches",
    "Deep learning models improve performance in NLP tasks"
]
```

```python
# -----------------------------
# 1. Bag-of-Words (Count Occurrence)
# -----------------------------
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(corpus)

print("Vocabulary:", count_vectorizer.get_feature_names_out())
print("\nBag-of-Words (Count Occurrence):")
print(pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out()))
```

```python
# -----------------------------
# 1. Bag-of-Words (Count Occurrence)
# -----------------------------
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(corpus)

print("Vocabulary:", count_vectorizer.get_feature_names_out())
print("\nBag-of-Words (Count Occurrence):")
print(pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out()))
```

```python
# -----------------------------
# 3. TF-IDF
# -----------------------------
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

print("\nTF-IDF Matrix:")
print(pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out()))
```

```python
# -----------------------------
# 4. Word2Vec Embeddings
# -----------------------------
# Tokenize corpus
tokenized_corpus = [word_tokenize(doc.lower()) for doc in corpus]

# Train Word2Vec
w2v_model = Word2Vec(sentences=tokenized_corpus, vector_size=50, window=5, min_count=1, workers=4)
```

```python
corpus1 = [
    "Natural language processing enables computers to understand human language",
    "Word embeddings like Word2Vec represent words in vector space",
    "Bag of words and TF-IDF are traditional NLP approaches",
    "Deep learning models improve performance in NLP tasks",
    "The king and queen ruled the kingdom while the man and woman worked in the village"
]
```

```python
# -----------------------------
# 1. Bag-of-Words (Count Occurrence)
# -----------------------------
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(corpus1)

bow_df = pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())
print("\n=== Bag-of-Words ===")
print(bow_df)
```

```python
# -----------------------------
# 2. Normalized Term Frequency
# -----------------------------
normalized_count = count_matrix.toarray().astype(float)
normalized_count = normalized_count / normalized_count.sum(axis=1, keepdims=True)

tf_df = pd.DataFrame(normalized_count, columns=count_vectorizer.get_feature_names_out())
print("\n=== Normalized Term Frequency ===")
print(tf_df)
```

```python
# -----------------------------
# 3. TF-IDF
# -----------------------------
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus1)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("\n=== TF-IDF ===")
print(tfidf_df)
```

```python

# -----------------------------
# 4. Word2Vec Embeddings
# -----------------------------
# Tokenize corpus
tokenized_corpus = [word_tokenize(doc.lower()) for doc in corpus1]

# Train Word2Vec
w2v_model = Word2Vec(sentences=tokenized_corpus, vector_size=50, window=5, min_count=1, workers=4)
```

```python
# -----------------------------
# 5. Explore Word Embeddings
# -----------------------------
words = ["king", "queen", "man", "woman", "computer"]

# (a) Word Embeddings
embeddings = {w: w2v_model.wv[w] for w in words if w in w2v_model.wv}
emb_df = pd.DataFrame(embeddings).T
print("\n=== Word2Vec Embeddings (first 5 dims) ===")
print(emb_df.iloc[:, :5])  # show only first 5 dimensions for readability

# (b) Most Similar Words
similar_words = {w: w2v_model.wv.most_similar(w, topn=5) for w in embeddings.keys()}
sim_df = pd.DataFrame({k: [x[0] for x in v] for k, v in similar_words.items()})
print("\n=== Most Similar Words ===")
print(sim_df)

# (c) Vector Arithmetic: king - man + woman ≈ queen
if all(word in w2v_model.wv for word in ["king", "man", "woman"]):
    result = w2v_model.wv.most_similar(positive=["king", "woman"], negative=["man"], topn=5)
    print("\n=== Vector Arithmetic (king - man + woman) ===")
    print(result)
```

```python
# -----------------------------
# 6. Visualization (optional)
# -----------------------------
# Plot embeddings of chosen words (2D via PCA)
from sklearn.decomposition import PCA

chosen_words = list(embeddings.keys())
vectors = [embeddings[w] for w in chosen_words]

pca = PCA(n_components=2)
reduced = pca.fit_transform(vectors)

plt.figure(figsize=(6,6))
plt.scatter(reduced[:,0], reduced[:,1])

for i, word in enumerate(chosen_words):
    plt.annotate(word, (reduced[i,0], reduced[i,1]))

plt.title("Word2Vec Embeddings (PCA 2D Projection)")
plt.show()
```
