```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# Download required resources
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
# Sample text
text = "Steve Jobs founded Apple Inc. in California. He was born in San Francisco."
```

```python
# Tokenization and POS tagging
tokens = word_tokenize(text)
tags = pos_tag(tokens)
```

```python
# Named Entity Recognition
named_entities = ne_chunk(tags)
```

```python
# Display result
print(named_entities)
```

## Student Activity:
1.Perform NER on a paragraph from a newspaper or academic article.

```python
newText = """Apple Inc. announced on Monday that it will invest $1 billion 
to expand its operations in Austin, Texas. CEO Tim Cook said the expansion 
will create around 5,000 new jobs in the technology sector. 
Meanwhile, President Joe Biden praised the investment as a boost 
to the U.S. economy."""
```

```python
# Tokenization and POS tagging 
tokens = word_tokenize(newText) 
tags = pos_tag(tokens) 
named_entities = ne_chunk(tags) 
print(named_entities)
```

2.Count how many entities were identified for each type (e.g., PERSON, ORGANIZATION, and LOCATION).

```python
# Extract entity types and count them
entity_counts = Counter()
for chunk in named_entities:
    if hasattr(chunk, 'label'):  # Named entity detected
        entity_counts[chunk.label()] += 1
```

```python
# Display results
print("Named Entity Counts:")
for entity_type, count in entity_counts.items():
    print(f"{entity_type}: {count}")
```

```python
# --- Visualization ---
plt.figure(figsize=(8,5))
plt.bar(entity_counts.keys(), entity_counts.values())
plt.xlabel("Entity Type")
plt.ylabel("Frequency")
plt.title("Named Entity Recognition (NER) Counts")
plt.show()
```
