```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# Download necessary data
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Sample text
text = """Natural Language Processing is an exciting field. It includes tokenization, stemming, and removing stop words."""
```

```python
# Tokenization
sentences = sent_tokenize(text)
words = word_tokenize(text)
print("Sentences:", sentences)
print("Words:", words)
```

```python
# Stop word removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
print("Filtered Words:", filtered_words)
```

```python
# Stemming
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]
```

```python
print ("Stemmed Words:", stemmed_words)
```

```python
paragraph = """
Football is a popular sport worldwide. It improves physical health and builds teamwork skills.
Many schools encourage students to play football regularly.
"""
```

```python
# Tokenization
tokens = word_tokenize(paragraph)
stop_words = set(stopwords.words('english'))
```

```python
# Separate stop words and non-stop words
stop_words_in_text = [word for word in tokens if word.lower() in stop_words]
non_stop_words = [word for word in tokens if word.lower() not in stop_words]

# Stemming
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in non_stop_words]

# Output
print("Original Tokens:", tokens)
print("Stop Words Found:", stop_words_in_text)
print("Number of Stop Words:", len(stop_words_in_text))
print("Stemmed Words:", stemmed_words)
print("Number of Stemmed Words:", len(stemmed_words))
```
